```python
df = pd.read_csv("~/Desktop/state_cases_over_time.csv")
#df=df.dropna()
df
```

```python

confirmed=df['confirmed'].values
deaths=df['deaths'].values
plt.plot(confirmed,deaths)
plt.xlabel('confirmed')
plt.ylabel('deaths')
plt.title('Covid Deaths')
plt.grid(True)
plt.savefig("Covid.png")
plt.show()
```

```python
df2=df[['province_state','confirmed','deaths','country_region']].dropna()
df2=df2[df2['country_region']=='US']
gb=df2.groupby(by='province_state').sum()

x=gb.index
confirmed=gb['confirmed']
deaths=gb['deaths']
plt.bar(x,confirmed,color='blue',width=3)
plt.show()
```
